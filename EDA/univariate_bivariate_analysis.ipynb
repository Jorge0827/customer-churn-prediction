```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
#¿Cuanto tiempo llevan los clientes?
#¿La mayoría son nuevos o antiguos?
sns.histplot(df["tenure"])
```

```python
sns.countplot(x="InternetService", data=df)
```

```python
sns.countplot(x="StreamingMovies", data=df)
```

```python
sns.countplot(x="Churn", data=df)

plt.title("Distribución de churn")
plt.show()
```

```python
sns.countplot(x="Contract", hue="Churn", data=df)
#Entre contrato más largo menos churn
#Contrato mensual más tendencia a ser churn
```

```python
sns.boxplot(x="Churn", y="tenure", data=df)
#Clientes con churn tienen tenure más bajo.
```

```python
#Correlaciones
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True)
```
